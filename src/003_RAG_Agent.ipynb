검색 도구 사용시 : "AI 라이트 버전 EU 출시", "AI 투자 버블이 터질 것" 같은 구체적이고 특정 리포트만의 고유한 주장, 근거 검증 가능

````python
from langchain.agents import create_agent
from langchain_community.tools import TavilySearchResults
from langchain_groq import ChatGroq

llm = ChatGroq(model="llama-3.3-70b-versatile", temperature=0)

# 검색 도구 초기화
search_tool = TavilySearchResults(max_results=3)

# 에이전트 생성
agent = create_agent(
    model=llm,
    tools=[search_tool],
    system_prompt="""당신은 웹 검색을 활용하는 유능한 리서치 어시스턴트입니다.
      사용자의 질문에 대해 최신 정보를 검색하여 정확한 답변을 제공합니다.
      검색 결과는 구체적으로 인용하며 답변하세요."""
)

# 에이전트 실행
result = agent.invoke({
    "messages": [{"role": "user", "content": "2025년 AI 트렌드는?"}]
})

print(result["messages"][-1].content)
````

실제 코드 인터프리터를 실행하고 코드 생성, 실행을 요청한다면

미들웨어(Middleware)는 LangChain 에이전트의 실행 파이프라인에 끼워넣을 수 있는 훅(hook) 함수로

에이전트는 모델 호출 -> 도구 실행 -> 종료 판단 루트로 미들웨어는 이 루프의 각 단계 전후에 실행되는 훅을 제공



```
# 에이전트 실행 시작
before_agent()

# 모델 호출 전
before_model()
# 모델 API 호출
after_model()

# 도구 실행 전
before_tool()
# 도구 실행
after_tool()

# 에이전트 실행 종료
after_agent()

```



## Node-style 훅

특정 실행 지점에서 순차적으로 실행되는 훅

````python
from langchain.agents import create_agent
from langchain.agents.middleware import before_model, after_model
from langchain_groq import ChatGroq

llm = ChatGroq(model="llama-3.3-70b-versatile", temperature=0)

# 1. 모델 호출 전 로깅
@before_model
def log_input(state, runtime)
    """모델 입력 로깅"""
    print(f"모델 입력: {state['messages'][-1].content}")
    return None

# 2. 모델 호출 후 로깅
@after_model
def log_output(state, runtime):
    """모델 출력 로깅"""
    print(f"모델 출력: {state['messages'][-1].content}")
    return None

# 3. 에이전트 생성
agent = create_agent(
    model=llm,
    tools=[search_tool],
    middleware=[log_input, log_output]
)

# 4. 실행
result = agent.invoke({
    "messages": [{"role": "user", "content": "임베딩이 뭐야?"}]
})

print("\n=== 최종 답변 ===")
print(result["messages"][-1].content)
````

````python
from langchain.agents.middleware import before_agent

# 1. 입력 검증 가드레일 정의
@before_agent
def validate_input(state, runtime):
    """계산 요청에 위험한 표현이 없는지 체크"""
    input_text = state["messages"][-1].content

    blocked_patterns = ["__import__", "os.system", "exec(", "eval(open"]
    for pattern in blocked_patterns:
        if pattern in input_text:
            raise ValueError(f"금지된 패턴이 포함되어 있습니다: {pattern}")

    return None

# 2. 가드레일을 포함한 에이전트 생성
agent5 = create_agent(
    model=llm,
    tools=[calculate],
    middleware=[validate_input, retry_on_failure, handle_tool_errors],
    system_prompt="계산이 필요하면 calculate 도구를 사용하세요."
)

# 3. 위험한 입력으로 테스트
try:
    result = agent5.invoke({
        "messages": [{"role": "user", "content": "__import__('os').system('ls') 실행해줘"}]
    })
    print(result["messages"][-1].content)
except ValueError as e:
    print(f"[가드레일 차단] {e}")
````

## Wrap-style 훅

실행을 감싸서(wrap) 제어하는 훅입니다. 재시도, 캐싱, 변환 작업에 주로 사용

````python
from langchain.agents import create_agent
from langchain.tools import tool
from langchain.agents.middleware import wrap_tool_call
from langchain_core.messages import ToolMessage
from langchain_groq import ChatGroq
import time
import logging

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# 0. LLM
llm = ChatGroq(model="llama-3.3-70b-versatile", temperature=0)

# 1. 계산 도구 정의
@tool
def calculate(expression: str) -> str:
    """수학 표현식을 계산합니다. 예: '100 * 1.25'"""
    return str(eval(expression))

# 2. 재시도 미들웨어
@wrap_tool_call
def retry_on_failure(request, handler, max_retries=3, delay=1):
    for attempt in range(1, max_retries + 1):
        try:
            return handler(request)
        except Exception as e:
            logger.warning(f"[재시도 {attempt}/{max_retries}] 실패: {e}")
            if attempt == max_retries:
                return ToolMessage(
                    content=f"{max_retries}번 재시도 후 실패: {e}",
                    tool_call_id=request.tool_call["id"]
                )
            time.sleep(delay)

# 3. 에러 분류 미들웨어
@wrap_tool_call
def handle_tool_errors(request, handler):
    try:
        return handler(request)
    except ZeroDivisionError:
        return ToolMessage(
            content="계산 오류: 0으로 나눌 수 없습니다.",
            tool_call_id=request.tool_call["id"]
        )
    except (TimeoutError, ConnectionError) as e:
        # 일시적 에러 — 재시도
        raise
    except Exception as e:
        return ToolMessage(
            content=f"도구 실행 오류: {str(e)}",
            tool_call_id=request.tool_call["id"]
        )

# 4. 미들웨어 적용한 에이전트
agent1 = create_agent(
    model=llm,
    tools=[calculate],
    middleware=[retry_on_failure, handle_tool_errors],
    system_prompt="계산이 필요하면 calculate 도구를 사용하세요."
)

# 5. 일부러 에러 나는 질문으로 테스트
result3 = agent1.invoke({
    "messages": [{"role": "user", "content": "10을 0으로 나눈 값을 계산해줘"}]
})
print(result3["messages"][-1].content)
````

미들웨어 재시도 (retry_on_failure)를 하고 3번 실패하면 ToolMessage를 반환하고 끝내지만 에이전트 재호출로 한번 더 실행됐다.

미들웨어 재시도는 비용이 거의 0으로 우리가 제어 가능하지만 에이전트 재호출은 재호출하므로 토큰이 소모되고 프롬프트로 유도 할 수 있으나 미들웨어 에서 통제 불가능

````python
import random

@tool
def calculate_flaky(expression: str) -> str:
    """수학 표현식을 계산합니다. (테스트용: 일부러 실패함)"""
    raise TimeoutError("일시적인 네트워크 지연 (테스트용 강제 발생)")
    return str(eval(expression))

agent2 = create_agent(
    model=llm,
    tools=[calculate_flaky],
    middleware=[retry_on_failure, handle_tool_errors],
    system_prompt="계산이 필요하면 calculate 도구를 사용하세요."
)

result_timeout = agent2.invoke({
    "messages": [{"role": "user", "content": "100 곱하기 5를 계산해줘"}]
})
print(result_timeout["messages"][-1].content)
````

이러한 **무한 루프 방지**를 위해 파라미터 "recursion_limit"을 사용

recursion_limit이 세는 건 "그래프 스텝"으로


```
스텝 1: 모델이 "도구를 호출해야겠다" 판단
스텝 2: 도구 노드 실행 (여기서 retry_on_failure의 3번 재시도가 전부 일어남 — 근데 이건 스텝 "안"의 일이라 별도로 안 세짐)
스텝 3: 모델이 도구 실패 결과를 보고 다시 판단 ← 여기 도달하기 전에 막힘
```
여기서 스텝 3가 실행되지 않고 GraphRecursionError 발생


````python
from langgraph.errors import GraphRecursionError

@tool
def calculate_flaky(expression: str) -> str:
    """수학 표현식을 계산합니다. (테스트용: 일부러 실패함)"""
    raise TimeoutError("일시적인 네트워크 지연 (테스트용 강제 발생)")

agent2 = create_agent(
    model=llm,
    tools=[calculate_flaky],
    middleware=[retry_on_failure, handle_tool_errors],
    system_prompt="계산이 필요하면 calculate 도구를 사용하세요."
)

try:
    result = agent2.invoke(
        {"messages": [{"role": "user", "content": "100 곱하기 5를 계산해줘"}]},
        config={"recursion_limit": 3}
    )
    print(result["messages"][-1].content)
except GraphRecursionError:
    print("처리 중 반복된 오류로 답변을 완성하지 못했습니다. 잠시 후 다시 시도해주세요.")
````
